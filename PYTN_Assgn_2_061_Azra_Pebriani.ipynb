* This dataset uses the financial definition of a building/building unit, for tax purposes. In case a single entity owns the building in question, a sale covers the value of the entire building. It looks like empty records are not being treated as NA. We convert columns to their appropriate data types to obtain NAs.

```python
#First, let's check which columns should be categorical
print('NEIGHBORHOOD')
for col in nyc.columns:
    if nyc[col].dtype=='object':
        print(col, nyc[col].nunique())
```

```python
# LAND SQUARE FEET,GROSS SQUARE FEET, SALE PRICE, BOROUGH should be numeric. 
# SALE DATE datetime format.
# categorical: NEIGHBORHOOD, BUILDING CLASS CATEGORY, TAX CLASS AT PRESENT, BUILDING CLASS AT PRESENT,
# BUILDING CLASS AT TIME OF SALE, TAX CLASS AT TIME OF SALE,BOROUGH 

numer = ['LAND SQUARE FEET','GROSS SQUARE FEET', 'SALE PRICE', 'BOROUGH']
for col in numer: # coerce for missing values
    nyc[col] = pd.to_numeric(nyc[col], errors='coerce')

categ = ['NEIGHBORHOOD', 'BUILDING CLASS CATEGORY', 'TAX CLASS AT PRESENT', 'BUILDING CLASS AT PRESENT', 'BUILDING CLASS AT TIME OF SALE', 'TAX CLASS AT TIME OF SALE']
for col in categ:
    nyc[col] = nyc[col].astype('category')

nyc['SALE DATE'] = pd.to_datetime(nyc['SALE DATE'], errors='coerce')
```

Our dataset is ready for checking missing values.

```python
missing = nyc.isnull().sum()/len(nyc)*100

print(pd.DataFrame([missing[missing>0],pd.Series(nyc.isnull().sum()[nyc.isnull().sum()>1000])], index=['percent missing','how many missing']))
```

Around 30% of GROSS SF and LAND SF are missing. Furthermore, around 17% of SALE PRICE is also missing.

We can fill in the missing value from one column to another, which will help us reduce missing values. Expected values:

(6, 20)

(1366, 20)

```python
print(nyc[(nyc['LAND SQUARE FEET'].isnull()) & (nyc['GROSS SQUARE FEET'].notnull())].shape)
print(nyc[(nyc['LAND SQUARE FEET'].notnull()) & (nyc['GROSS SQUARE FEET'].isnull())].shape)
```

There are 1372 rows that can be filled in with their approximate values.

```python
nyc['LAND SQUARE FEET'] = nyc['LAND SQUARE FEET'].mask((nyc['LAND SQUARE FEET'].isnull()) & (nyc['GROSS SQUARE FEET'].notnull()), nyc['GROSS SQUARE FEET'])
nyc['GROSS SQUARE FEET'] = nyc['GROSS SQUARE FEET'].mask((nyc['LAND SQUARE FEET'].notnull()) & (nyc['GROSS SQUARE FEET'].isnull()), nyc['LAND SQUARE FEET'])
```

## Harga Jual Properti di Manhattan dengan di Brooklyn berbeda secara signifikan atau tidak?


Population: Harga Jual Properti di Wilayah Manhattan & Brooklyn di dataset (nyc-rolling_sales.csv)
Parameter: 
- Harga Jual Properti di Manhattan
- Harga Jual Properti di Brooklyn

h1
Harga Jual Properti di Manhattan != Harga Jual Properti di Brooklyn

h0
Harga Jual Properti di Manhattan = Harga Jual Properti di Brooklyn

```python
nyc['Wilayah'] = nyc['BOROUGH'].replace({
    1:'Manhattan', 
    2:'Bronx', 
    3:'Brooklyn', 
    4:'Queens', 
    5:'Staten Island'})

nyc['Harga Jual'] = pd.to_numeric(nyc['SALE PRICE'], errors='coerce').convert_dtypes()

# nyc = nyc.dropna()
nyc.head()
```

```python
# Write your function below

manhattan = nyc[nyc['Wilayah'] == 'Manhattan']
brooklyn = nyc[nyc['Wilayah'] == 'Brooklyn']
```

```python
n_m = len(manhattan)
mean_m = manhattan['Harga Jual'].mean()
std_m = manhattan['Harga Jual'].std()

print("Populasi di Wilayah Manhattan: ",n_m)
print("Rata-rata Harga Penjualan di Wilayah Manhattan: ", mean_m)
print("STD di Wilayah Manhattan: ", std_m)
```

```python
n_b = len(brooklyn)
mean_b = brooklyn['Harga Jual'].mean()
std_b = brooklyn['Harga Jual'].std()

print("Populasi di Wilayah Brooklyn: ",n_b)
print("Rata-rata Harga Penjualan di Wilayah Brooklyn: ", mean_b)
print("STD di Wilayah Brooklyn: ", std_b)
```

```python
# Graded-Funtion Begin

print(("zstat", "p-value"))
sm.stats.ztest(manhattan["Harga Jual"].dropna(), brooklyn["Harga Jual"].dropna(), alternative='two-sided')

# Graded-Funtion End
```
